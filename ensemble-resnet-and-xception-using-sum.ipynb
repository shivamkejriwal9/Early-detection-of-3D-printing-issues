```python
submission = pd.read_csv('/kaggle/input/3d-printing-issue-detection-model/submission_Resnet101V2.csv')
```

```python
submission1 = pd.read_csv('/kaggle/input/3d-printing-issue-detection-model/submission_Xception.csv')
```

```python
predict = submission['has_under_extrusion']
predict1 = submission1['has_under_extrusion']
```

```python
predict_1 = np.copy(predict)
predict1_1 = np.copy(predict1)
```

```python
mask1 = predict_1>=0.975
predict_1[mask1] = 1
predict_1[~mask1] = 0
```

```python
mask2 = predict1_1>=0.985
predict1_1[mask2] = 1
predict1_1[~mask2] = 0
```

```python
submission['has_under_extrusion'] = predict_1
submission1['has_under_extrusion'] = predict1_1
```
